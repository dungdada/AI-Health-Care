# 실습과제: 환자 응급도 분류하기

응급실에 내원한 환자의 생체징후를 확인하여 응급도를 분류하고 적절한 처치 방향을 안내하는 프로그램을 작성해보세요.

## 요구사항

1. **응급도 분류 기준**
   - 즉시 처치: 의식이 없거나 수축기혈압 ≥ 180 또는 체온 ≥ 39.0
   - 응급 처치: 통증이 8점 이상 또는 호흡곤란
   - 준응급: 통증이 4-7점 또는 발열(37.5-38.9도)
   - 비응급: 위 조건에 해당하지 않는 경우

2. **처치 방향 및 대기시간**
   - 즉시 처치: "즉시 응급처치" / 0분
   - 응급 처치: "응급 처치 후 정밀검사" / 15분
   - 준응급: "약물 처치 후 전문의 진료" / 30분
   - 비응급: "일반 진료" / 60분

3. **출력 형식**
```
=== 환자 응급도 분류 결과 ===
환자명: 김환자 (45세)
생체징후: 혈압 160/90, 체온 38.2도, 통증 3점
의식상태: 의식 있음, 호흡곤란: 호흡 없음

🟡 분류 결과: 준응급
📋 처치 방향: 약물 처치 후 전문의 진료
⏰ 예상 대기시간: 30분
```

### 힌트

- `if-elif-else`문으로 응급도를 분류하세요
- `and`/`or` 연산자로 복합 조건을 처리하세요
- `not` 연산자로 의식 상태를 확인하세요
- f-string을 사용해 결과를 예쁘게 출력하세요

## 💻 구현해보세요

````python
# 기본 데이터 - 환자 정보
patient = {
    "name": "김환자",                      # 환자 이름
    "age": 45,                            # 환자 나이 (세)
    "systolic_bp": 160,                   # 수축기 혈압 (mmHg)
    "temperature": 38.2,                  # 체온 (섭씨)
    "pain_level": 3,                      # 통증 점수 (0-10점)
    "is_conscious": True,                 # 의식 상태 (True: 의식있음, False: 의식없음)
    "has_breathing_difficulty": False     # 호흡곤란 여부 (True: 있음, False: 없음)
}
````

````python
# 여기에 코드를 작성하세요
print("=== 환자의 응급도 분류 결과 ===")
# 1. 환자 기본 정보 출력
print(f"환자명 : {patient["name"]} ({patient["age"]}세)")
print(f"생체징후: 혈압 {patient["systolic_bp"]}/90, 체온 {patient["temperature"]}도, 통증 {patient["pain_level"]}점")
conscious = "의식있음" if patient["is_conscious"] else "의삭없음"
breathing = "호흡있음" if patient["has_breathing_difficulty"] else "호흡없음"

print(f"의식상태: {conscious}, 호흡곤란:{breathing}")
# 2. 응급도 분류 (if-elif-else 활용)


if not patient["is_conscious"] or patient["systolic_bp"]>=180 or patient["temperature"]>=39.0:
  emergency_level = "🔴즉시처치"
elif patient["pain_level"]>=8 or not patient["has_breathing_difficulty"]:
  emergency_level ="🟠응급처치"
elif 4<=patient["pain_level"]<=7 or 37.5<=patient["temperature"]<=38.9:
  emergency_level ="🟡준응급"
else:
  emergency_level="🟢비응급"
print(f"\n분류결과: {emergency_level}")


# 3. 처치 방향 및 대기시간 안내
if emergency_level=="🔴즉시처치":
  treatment="즉시 응급처치"
  treatment_time="0분"
elif emergency_level=="🟠응급처치":
  treatment="응급 처치 후 정밀검사"
  treatment_time=int(15)
elif emergency_level=="🟡준응급":
  treatment="약물 처치 후 전문의 진료"
  treatment_time=int(30)
else:
  treatment="일반 진료"
  treatment_time=int(60)

print(f"처치방향: 📋{treatment}")
print(f"예상 대기시간: ⏰{treatment_time}분")
````

````python
# @title 정답

# 환자 정보 (딕셔너리)
patient = {
    "name": "김환자",                      # 환자 이름
    "age": 45,                            # 환자 나이 (세)
    "systolic_bp": 160,                   # 수축기 혈압 (mmHg)
    "temperature": 38.2,                  # 체온 (섭씨)
    "pain_level": 3,                      # 통증 점수 (0-10점)
    "is_conscious": True,                 # 의식 상태 (True: 의식있음, False: 의식없음)
    "has_breathing_difficulty": False     # 호흡곤란 여부 (True: 있음, False: 없음)
}

# 1. 환자 기본 정보 출력
print("=== 환자 응급도 분류 결과 ===")
print(f"환자명: {patient['name']} ({patient['age']}세)")
print(f"생체징후: 혈압 {patient['systolic_bp']}/90, 체온 {patient['temperature']}도, 통증 {patient['pain_level']}점")

consciousness_status = "의식 있음" if patient["is_conscious"] else "의식 없음"
breathing_status = "호흡 있음" if patient["has_breathing_difficulty"] else "호흡 없음"
print(f"의식상태: {consciousness_status}, 호흡곤란: {breathing_status}")
print()

# 2. 응급도 분류
if not patient["is_conscious"] or patient["systolic_bp"] >= 180 or patient["temperature"] >= 39.0:
    urgency_level = "즉시 처치"          # 응급도 분류 결과
    urgency_icon = "🔴"                # 응급도 표시 아이콘
    treatment = "즉시 응급처치"           # 처치 방향
    wait_time = "0분"                  # 예상 대기시간

elif patient["pain_level"] >= 8 or patient["has_breathing_difficulty"]:
    urgency_level = "응급 처치"
    urgency_icon = "🟠"
    treatment = "응급 처치 후 정밀검사"
    wait_time = "15분"

elif (patient["pain_level"] >= 4 and patient["pain_level"] <= 7) or (patient["temperature"] >= 37.5 and patient["temperature"] < 39.0):
    urgency_level = "준응급"
    urgency_icon = "🟡"
    treatment = "약물 처치 후 전문의 진료"
    wait_time = "30분"

else:
    urgency_level = "비응급"
    urgency_icon = "🟢"
    treatment = "일반 진료"
    wait_time = "60분"

# 3. 결과 출력
print(f"{urgency_icon} 분류 결과: {urgency_level}")
print(f"📋 처치 방향: {treatment}")
print(f"⏰ 예상 대기시간: {wait_time}")
````
